# Logistic map

```python
import matplotlib.pyplot as plt

def logistic_map(a, x0, n_iter):
    """
    Computes the Logistic map.

    Parameters:
    a (float): The parameter 'a' controlling the map's behavior.
    x0 (float): The initial value of x (0 < x0 < 1).
    n_iter (int): Number of iterations.

    Returns:
    list: Values of x over the iterations.
    """
    x_values = [x0]
    x = x0

    for _ in range(n_iter):
        x = a * x * (1 - x)
        x_values.append(x)

    return x_values

# Parameters for the Logistic map
x0 = 0.5  # Initial value of x
n_iter = 1000  # Number of iterations to stabilize
n_last = 100  # Number of iterations to plot after stabilization

# Range of 'a' values
a_values = [i / 1000 for i in range(2500, 4000)]  # a from 2.5 to 4.0

# Prepare data for bifurcation diagram
a_plot = []
x_plot = []

for a in a_values:
    x_values = logistic_map(a, x0, n_iter)
    x_plot.extend(x_values[-n_last:])  # Take the last n_last values
    a_plot.extend([a] * n_last)  # Repeat 'a' for each x value

# Plot the bifurcation diagram
plt.figure(figsize=(12, 8))
plt.scatter(a_plot, x_plot, s=0.1, color="blue")
plt.title("Bifurcation Diagram of the Logistic Map", fontsize=14)
plt.xlabel("a", fontsize=12)
plt.ylabel("x", fontsize=12)
plt.grid(True, linestyle='-', alpha=0.5)
plt.show()
```
