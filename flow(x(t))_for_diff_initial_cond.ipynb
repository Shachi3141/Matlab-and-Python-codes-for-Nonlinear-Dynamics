# Flow or asymptotic behaviour of the equation 
$$ \dot x = rx + x^3 - x^5 $$
for different value of x0 and r.

**Topic:** Bifurcation

*Date:* `23-01-2025`

```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt 

t = np.linspace(0,100,500)
x0 = -0.6
r = -1/4
def f(x,t):
    return (r*x+x**(3)-x**(5))

soln = odeint(f,x0,t)

plt.plot(t,soln)
plt.title(" Solution")
plt.axhline(y=1/np.sqrt(2), color='blue', linestyle='--', linewidth=1.5)
plt.axhline(y=-1/np.sqrt(2), color='blue', linestyle='--', linewidth=1.5)
plt.xlabel("Time")
plt.ylabel("X")
plt.grid()
plt.show()
```

```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt

# Time range
t = np.linspace(0, 100, 500)
r = -1/4
# Define the function for the ODE
def f(x, t):
    return r*x + x**3 - x**5

# Initial conditions to solve for
initial_conditions = [0.1, 0.5,0.65, 1.0, 1.25, -0.1, -0.5,-0.65, -1.0, -1.25]

# Plotting
plt.figure(figsize=(10, 6))

for x0 in initial_conditions:
    # Solve the ODE for each initial condition
    soln = odeint(f, x0, t)
    # Plot the solution
    plt.plot(t, soln, label=f'x₀ = {x0}')

# Add horizontal lines for ±1/√2
plt.axhline(y=1/np.sqrt(2), color='blue', linestyle='--', linewidth=1.5, label='y = ±1/√2')
plt.axhline(y=-1/np.sqrt(2), color='blue', linestyle='--', linewidth=1.5)

# Customize the plot
plt.title("Solutions for Different Initial Conditions")
plt.xlabel("Time")
plt.ylabel("X")
plt.legend()
plt.grid()
plt.show()
```

### for $r = -1/4$, $x=0$ is stable fixed point and $ x= \pm \frac{1}{\sqrt{2}} $ is a saddle fixed point
